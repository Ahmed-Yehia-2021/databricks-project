# Intializing Paths

```python
raw_resources = "/Volumes/dev_workspace/bronze/raw_resources/"
files = ["source_crm/cust_info.csv", "source_crm/prd_info.csv", "source_crm/sales_details.csv",     
         "source_erp/CUST_AZ12.csv", "source_erp/LOC_A101.csv", "/source_erp/PX_CAT_G1V2.csv"]
```

```python
workspace_path = "dev_workspace.bronze."
tables = ["crm_cust_info", "crm_prd_info", "crm_sales_details",
          "erp_cust_az12", "erp_loc_a101", "erp_px_cat_g1v2"]
```

# Loading The Data

```python

for i in range(6):
    df = spark.read.option("header", "true").option("inferschema", "true").csv(raw_resources+files[i])
    df.write.mode("overwrite").saveAsTable(workspace_path+tables[i])
```
